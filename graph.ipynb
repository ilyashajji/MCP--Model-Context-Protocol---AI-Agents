```python
from langchain.agents import create_agent
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
from langchain_mcp_adapters.client import MultiServerMCPClient
from langchain.messages import HumanMessage
from IPython.display import Markdown
```

```python
load_dotenv(override=True)
```

```python
mcp_client = MultiServerMCPClient({
    "mcp-server":{
        "transport":"streamable_http",
        "url":"http://localhost:24000/mcp"
}})
```

```python
tools=await mcp_client.get_tools()
```

```python
tools
```

```python
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)
```

```python
agent =create_agent(
    model=llm,tools=tools,system_prompt="answer user question using provided tools"
)
```

```python
resp = await agent.ainvoke({
    "messages":[
        HumanMessage("quel est le salire de hassan")
    ]
})
```

```python
print(resp['messages'][-1].content)
```

```python
resp = await agent.ainvoke({
    "messages":[
        HumanMessage("Quel est l outils le plus puissant en 2026 dans l ecommerce")
    ]
})
```

```python
print(display(Markdown(resp['messages'][-1].content)))
```
